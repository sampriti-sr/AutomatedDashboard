```python
# Define the marking scheme (mapping cleaned question names to scores)
marking_scheme = {
    "Q1: Which of the following best identifies you?": {"Option A": 0, "Option B": 1, "Option C": 2, "Option D": -1},
    "Q2: Which of the following scenarios best reflects your sense of achievement?": {"Option A": -1, "Option B": 2, "Option C": 1, "Option D": 0},
    "Q3: Which of the following best describes how you handle decision-making in your daily life?": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Q4: Which of the following best describes how you handle situations that require patience?": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Q5: Which of the following best describes how you view and understand yourself?": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Situation1: You are offered an internship with a decision-maker in a startup. The internship involves working on an ambiguous project with minimal guidance. Your primary goal is to gain valuable experience, but you also have upcoming placement interviews.": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Situation2: You are given a challenging project in your internship that requires solving a complex problem with no clear instructions. The success of the project depends on how you approach and break down the problem.": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Situation3: Your college is offering a semester-long elective course that will require significant time and effort but offers no immediate grade or credit benefit. How do you decide whether to enroll?": {"Option A": 2, "Option B": 1, "Option C": -1, "Option D": 0},
    "Situation4: You are offered an internship that provides significant hands-on experience and learning opportunities, but it does not come with an official certificate. You have the chance to work closely with decision-makers on real projects.": {"Option A": 1, "Option B": -1, "Option C": 2, "Option D": 0},
    "Situation5: You are involved in an internship project that requires ongoing effort and iteration. The project will not show immediate results, and it requires a focus on the process rather than expecting quick outcomes.": {"Option A": 2, "Option B": -1, "Option C": 1, "Option D": 0},
}

# Map original column names to cleaned ones
column_mapping = {
    "  Q1: Which of the following best identifies you?  ": "Q1: Which of the following best identifies you?",
    "Q2: Which of the following scenarios best reflects your sense of achievement?  ": "Q2: Which of the following scenarios best reflects your sense of achievement?",
    "Q3: Which of the following best describes how you handle decision-making in your daily life? ": "Q3: Which of the following best describes how you handle decision-making in your daily life?",
    "Q4: Which of the following best describes how you handle situations that require patience? ": "Q4: Which of the following best describes how you handle situations that require patience?",
    "Q5: Which of the following best describes how you view and understand yourself? ": "Q5: Which of the following best describes how you view and understand yourself?",
}

# Rename columns for consistency
df.rename(columns=column_mapping, inplace=True)

# Initialize a dictionary to store scores for each question
scores = {col: [] for col in marking_scheme.keys()}

# Process each row in the DataFrame
for _, row in df.iterrows():
    for question, options in marking_scheme.items():
        # Extract the response for the current question
        response = row[question].strip()  # Strip any extra whitespace
        # Append the score based on the response
        scores[question].append(options.get(response, 0))  # Default score is 0 for invalid options

# Add scores as new columns to the DataFrame
for question, score_list in scores.items():
    df[f"{question}_Score"] = score_list
```

```python
# Calculate the total score for each candidate
df["Total_Score"] = df[[f"{q}_Score" for q in marking_scheme.keys()]].sum(axis=1)

# Save the updated DataFrame to a new CSV file
df.to_csv("Updated_Advanced_Screening_Scores.csv", index=False)

print("The updated dataset with scores has been saved as 'Updated_Advanced_Screening_Scores.csv'")
```
